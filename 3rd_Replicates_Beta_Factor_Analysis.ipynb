```python
import vmd
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
sns.set_context('poster')
```

```python

#Calculation of contacts



mol = vmd.molecule.load('pdb','npt_dnmt3a_with_procaine_3rd.pdb')

n_frames=vmd.molecule.numframes(mol)

contact = []

#LIG=procaine

variant = 'dnmt3a_lig'
for frame in range(0, n_frames):
    print('Reading ' + str(frame) + 'th frame..')
    pdb = vmd.molecule.set_frame(mol, frame)

    dnmt3 = vmd.atomsel('protein and noh')
    dnmt3_protein_first = dnmt3.resid[0]
    dnmt3_protein_last = dnmt3.resid[-1]

    for j in range(dnmt3_protein_first, 913):
        dnmt3a_protein = vmd.atomsel(molid=mol, selection=f'protein and resid {j} and noh')


        dnmt3a_lig = vmd.atomsel(molid=mol, selection='resname LIG')


        contact_lig = len(dnmt3a_protein.contacts(selection=dnmt3a_lig, cutoff=5)[0])

        contact_sah = 0





        contact.append([variant, j, contact_sah, contact_lig])

df = pd.DataFrame(contact, columns=['Variant', 'Residue', 'Contact_Sah', 'Contact_Lig'])
df.to_csv('contacts_npt_dnmt3a_with_procaine_2.csv', index=False)
```

```python

#Average of contacts


avg = []

for i in np.unique(df['Residue']):

    df3 = df[(df['Residue'] == i)]
    avg.append([i,df3['Contact_Lig'].mean()])

        
        

```

```python


#Beta-factor


from vmd import molecule
mol4 = vmd.molecule.load('pdb', 'init_dnmt3a_with_procaine.pdb')




for i in range(0,len(avg)):

    res = vmd.atomsel('chain D and resid ' + str(avg[i][0])) 
    
    res.beta = avg[i][1]
        

molecule.write(mol4, "pdb", "dnmt3a_procaine_beta_3rd.pdb", first=0)
```

```python

#Calculation of contacts



mol2 = vmd.molecule.load('pdb','npt_dnmt3a_with_cofactor_procaine_3rd.pdb')

n_frames=vmd.molecule.numframes(mol2)

contact2 = []

#LIG=procaine
#CFC=cofactor

variant = 'dnmt3a_sah_lig_env'
for frame in range(0, n_frames):
    print('Reading ' + str(frame) + 'th frame..')
    pdb = vmd.molecule.set_frame(mol2, frame)

    dnmt3 = vmd.atomsel('protein and noh')
    dnmt3_protein_first = dnmt3.resid[0]
    dnmt3_protein_last = dnmt3.resid[-1]

    for j in range(dnmt3_protein_first, 913):
        dnmt3a_protein = vmd.atomsel(molid=mol2, selection=f'protein and resid {j} and noh')


        dnmt3a_lig = vmd.atomsel(molid=mol2, selection='resname LIG')
        dnmt3a_sah = vmd.atomsel(molid=mol2, selection='resname CFC')

        contact_lig = len(dnmt3a_protein.contacts(selection=dnmt3a_lig, cutoff=5)[0])

        contact_sah = len(dnmt3a_protein.contacts(selection=dnmt3a_sah, cutoff=5)[0])





        contact2.append([variant, j, contact_sah, contact_lig])

df2 = pd.DataFrame(contact2, columns=['Variant', 'Residue', 'Contact_Sah', 'Contact_Lig'])
df2.to_csv('contacts_npt_dnmt3a_with_cofactor_procaine_2.csv', index=False)
```

```python

#Average of contacts



avg_2 = []

for i in np.unique(df2['Residue']):

    df4 = df2[(df2['Residue'] == i)]
    avg_2.append([i,df4['Contact_Lig'].mean()])

        
        
```

```python


#Beta-factor



from vmd import molecule
mol5 = vmd.molecule.load('pdb', 'init_dnmt3a_with_cofactor_procaine.pdb')




for i in range(0,len(avg_2)):

    res_2 = vmd.atomsel('chain D and resid ' + str(avg_2[i][0])) 
    
    res_2.beta = avg_2[i][1]
        

molecule.write(mol5, "pdb", "dnmt3a_procaine_cofactor_beta_3rd.pdb", first=0)
```
